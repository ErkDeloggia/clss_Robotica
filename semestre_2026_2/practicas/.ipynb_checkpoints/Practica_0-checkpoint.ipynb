## Previo de la Práctica 0

Que es ROS 2

Un paquete es una unidad organizativa para tu codigo de Ros 2. Si se quiere poder instalar un codigo o compartirlo con otros, entonces
necesitarás que este organizado en un paquete. Con los paquetes puedes liberar tu trabajo ROS 2 y permitir que otros lo construyan y usen
facilmente.

La creacion de paquetes en ROS 2 usa <amen> como sistema de compilacion y <colcon> como herramienta de compilación. Puedes crear un paquete
usando Cmake o Phyton, que son oficialmente compatibles, aunque existen otros tipos de compilaciones.

La estructura de paquetes en ROS 2 es modular, organizada y centrada en el directorio src de un workspace.

Qué compone un paquete de ROS 2

Los paquetes ROS 2 en Phyton y CMake tiene cada uno su contenido mínimo requerido:

+Python:

- package.xml Contiene metainformacion sobre el paquete.
- resource/<package_name> marcador del paquete.
- setup.cfg es necesario cuando un paquete tiene ejecutables.
- setup.py contiene las instrucciones sobre como instalar el paquete.
- <package_name> es un directorio con el mismo nombre que el paquete.
    
+CMake:

- CMakeLists.txt que describe cómo construir el código dentro del paquete.
- include/<package_name> Directorio que contiene los encabezados públicos del paquete.
- package.xml archivo que contiene metainformación sobre el paquete.
- src Directorio que contiene el código fuente del paquete.

Estos paquetes pueden estar junto con carpetas para código fuente, cabeceras, recursos, pruebas y archivos de lanzamiento.
Un solo espacio de trabajo puede contener tantos paquetes como quieras, cada uno en su propia carpeta. También puedes tener 
paquetes de diferentes tipos de compilación en un mismo espacio de trabajo (CMake, Python, etc.). No puedes tener paquetes anidados.
La mejor práctica es tener una carpeta dentro de tu espacio de trabajo y crear tus paquetes ahí <.src> . Esto mantiene el nivel superior 
del espacio de trabajo "limpio".

Definicion de Nodo

Un nodo es la unidad unidad básica de ejecución y procesamiento. Se trata de un programa ejecutable independiente 
(escrito generalmente en C++ o Python) que realiza una tarea específica dentro del sistema. Los nodos suelen ser 
una combinación compleja de editores, suscriptores, servidores de servicio, clientes de servicio, servidores de 
acción y clientes de acción, todo al mismo tiempo.

Mediante el protocolo middleware de comunicación DDS, en ROS 2 cada nodo tiene la capacidad de descubrir cualquier otro nodo, 
lo que da lugar a sistemas totalmente distribuidos.

Cada nodo en ROS debe ser responsable de un único propósito modular, por ejemplo, controlar los motores de ruedas 
o publicar los datos del sensor desde un telémetro láser. Cada nodo puede enviar y recibir datos de otros nodos a 
través de topicos, servicios, acciones o parámetros.

Un sistema robótico completo está compuesto por muchos nodos que trabajan en conjunto. En ROS 2, un único ejecutable 
(programa C++, programa Python, etc.) puede contener uno o más nodos.


  ![Nodo](imagenes_p0/Nodo.png)


Definición de Topic

Los topics son un método principal para transferir datos entre nodos, permitiendo la comunicación entre diferentes partes del sistema. 
ROS 2 descompone sistemas complejos en muchos nodos modulares. Los topics son un elemento vital del grafo ROS que actúan como un bus 
para que los nodos intercambien mensajes. Un nodo puede publicar datos sobre cualquier número de temas y simultáneamente tener suscripciones 
a cualquier número de temas.

Los topics son una de las principales formas en que los datos se mueven entre nodos y, por tanto, entre diferentes partes del sistema.

Nodo publicador:

#!urs/bin/env python3

import rclpy
from rclpy.node import  Node
from std_msgs.msg import Float64
import math
import time


class RpmNodePub(Node):
    def __init__(self):
        super().__init__("rpm_sin_publisher")

        #Publicador
        self.publicador_ = self.create_publisher(
            msg_type = Float64,
              topic ="topic_rpm",
                qos_profile= 10)
        self.timer_ = self.create_timer(
            timer_period_sec= 1.0, callback= self.cbck)

        self.get_logger().info("Nodo publicador de RPM Activado")

        #parametro de la onda senoidal
        
        self.star_time = time.time()
        self.amplitud = 1.0
        self.frequency = 0.1
        
        
    def cbck(self):
        msg = Float64()
        elapsed_time = time.time() - self.star_time


        #Funcion senoida 

        rpm_sine = self.amplitud * math.sin(2 * math.pi * self.frequency * elapsed_time)
        

        msg.data = abs(rpm_sine)
        self.publicador_.publish(msg)

        


def main(args= None):
    rclpy.init(args=args)
    node = RpmNodePub()
    rclpy.spin(node)
    rclpy.shutdown()

if __name__ == '__main__':
    main()


Nodo Subscriptor-Publicador:

#!urs/bin/env python3

import rclpy
from rclpy.node import  Node
from std_msgs.msg import Float64
import math
import time

class NodeSubPubRads(Node):
    def __init__(self):
        super().__init__("rpm_to_rads_converter_node")

        rpm = 0

        #Suscriptor

        self.create_subscription(
            msg_type= Float64,
                    callback= self.sub_cbck,
                        topic="topic_rpm", 
                            qos_profile= 10)
        
        #Publicador
        
        self.publisher_counter_ = self.create_publisher(
            msg_type= Float64,
            topic='topic_rads',
              qos_profile= 10
        )

        self.get_logger().info("Conversor de RPM a Rad/s Activado")
        
    def sub_cbck(self, msg):
        rpm = msg.data

        #Conversion de RPM a Rad/s

        rads = rpm  * ((2 * math.pi)/60)


        new_msg = Float64()
        new_msg.data = float(rads)
        self.publisher_counter_.publish(new_msg)


def main(args= None):
    rclpy.init(args=args)
    node = NodeSubPubRads()
    rclpy.spin(node)
    rclpy.shutdown()

if __name__ == '__main__':
    main()
